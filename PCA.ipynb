```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```
data = cancer['data']
target = cancer['target'].reshape(-1,1)
```

```
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled = scaler.fit_transform(data)
print(scaled)
```

```
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
P1_skln = pca.fit_transform(scaled)
print("original shape:   ", scaled.shape)
print("transformed shape:", P1_skln .shape)
```

```
plt.scatter(scaled[:, 0], scaled[:, 1],
            c=cancer.target, edgecolor='none',
            cmap=plt.cm.get_cmap('viridis', 10), 
            alpha=0.5)
plt.xlabel('component 1')
plt.ylabel('component 2')
plt.colorbar();
```

```
pca = PCA().fit(scaled)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```


It looks like for our data that we have here, anything over 6 or 7 components gets us over 90% varience.
